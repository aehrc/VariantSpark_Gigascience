```python
import numpy as np
import pandas as pd
```

```python
allPheno = pd.read_csv('cnf-01-run-1.pheno.csv', index_col=0).drop(['lbl'], axis=1)
allPheno
```

```python
print(allPheno.shape)
for exp in range(1,10):
    for run in range(1,4):
        name = "cnf-"+str(exp).zfill(2)+"-run-"+str(run)
        fname= name+".pheno.csv"
        print(fname)
        allPheno = allPheno.join(pd.read_csv(fname, index_col=0).rename(columns={'lbl': name}))
        print(allPheno.shape)
allPheno
```

```python
allPheno = allPheno.replace([0, 1], [False, True])
allPheno
```

```python
allPheno.to_csv('allPheno.csv')
```
